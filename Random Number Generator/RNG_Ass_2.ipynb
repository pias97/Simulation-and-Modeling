```python
import matplotlib.pyplot as plt
```

```python
def RN(val):
  initial_seed = int(val)
  RN = []
  for i in range(10000):
    next_seed = initial_seed**2
    new_seed = next_seed%1000000
    initial_seed = new_seed
    temp = new_seed/1000000
    #repeat
    for j in RN:
      if j == temp:
        print("Repeating random numbers. Process halted.")
        return RN  
    RN.append(temp)
  return RN
```

```python
RNarray = RN(123456)
print(RNarray)
print(len(RNarray))
```

```python
plt.hist(RNarray, bins = 200)
plt.show()
```
